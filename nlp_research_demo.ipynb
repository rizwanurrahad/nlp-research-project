# NLP Research Demo Notebook

This notebook demonstrates a small NLP workflow for application portfolio use.

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

texts = [
    "folklore narrative oral tradition",
    "cultural story village legend",
    "computer architecture operating system",
    "data structures algorithms compiler",
]
labels = [1, 1, 0, 0]

pipe = Pipeline([
    ("tfidf", TfidfVectorizer(lowercase=True, ngram_range=(1,2))),
    ("clf", LogisticRegression(max_iter=1000, solver="liblinear")),
])
pipe.fit(texts, labels)
print("Training complete")
```

```
test_texts = [
    "traditional bengali tale and oral literature",
    "distributed systems and network protocols",
]
preds = pipe.predict(test_texts)
for t, p in zip(test_texts, preds):
    print(p, "->", t)
```

## Error analysis example
A likely failure case is domain shift: a text may discuss culture using technical vocabulary, or vice versa. In a larger system I would inspect false positives/false negatives, add more Bengali/domain-specific samples, and compare TF-IDF against transformer embeddings.
